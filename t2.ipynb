```python
df['AP_COIDADE'] = pd.to_numeric(df['AP_COIDADE'], errors='coerce')
df['AP_NUIDADE'] = pd.to_numeric(df['AP_NUIDADE'], errors='coerce')

def calcular_idade(row):
    if pd.isna(row['AP_COIDADE']) or pd.isna(row['AP_NUIDADE']):
        return None
    coidade = row['AP_COIDADE']
    nuidade = row['AP_NUIDADE']
    if coidade == 1:  # Anos
        return nuidade
    elif coidade == 2:  # Meses
        return nuidade / 12
    elif coidade == 3:  # Dias
        return nuidade / 365
    elif coidade == 4:  # Anos (padrão para a maioria dos casos)
        return nuidade
    else:  # Outros códigos (ex.: 5 ou inválidos)
        return None

df['idade'] = df.apply(calcular_idade, axis=1)
df = df[df['idade'] < 18].copy()  
def calcular_faixa_etaria(idade):
    if pd.isna(idade) or idade < 0:
        return "Desconhecido"
    elif 0 <= idade <= 5:
        return "0-5 anos"
    elif 6 <= idade <= 12:
        return "6-12 anos"
    elif 13 <= idade <= 17:
        return "13-17 anos"
    return "Fora do Escopo"

df['faixa_etaria'] = df['idade'].apply(calcular_faixa_etaria)
df['ANO'] = df['AP_CMP'].astype(str).str[:4]
```

```python
import pandas as pd
import streamlit as st
import plotly.express as px  # Adicionando a importação de plotly.express

# Transformar em numérico
df['AP_NUIDADE'] = pd.to_numeric(df['AP_NUIDADE'], errors='coerce')
df['AP_COIDADE'] = pd.to_numeric(df['AP_COIDADE'], errors='coerce')

# Calcular idade com base nas unidades possíveis
def calcular_idade(row):
    if pd.isna(row['AP_COIDADE']):
        return None
    if row['AP_NUIDADE'] == 1:  # Anos
        return row['AP_COIDADE']
    elif row['AP_NUIDADE'] == 2:  # Meses
        return row['AP_COIDADE'] / 12
    elif row['AP_NUIDADE'] == 3:  # Dias
        return row['AP_COIDADE'] / 365
    else:  # Assumir anos para outros valores ou NaN
        return row['AP_COIDADE']

df['idade'] = df.apply(calcular_idade, axis=1)

# Filtrar apenas pacientes com menos de 18 anos
df = df[df['idade'] < 18]

# Calcular faixa_etaria dinamicamente
def calcular_faixa_etaria(idade):
    if pd.isna(idade) or idade < 0:
        return "Desconhecido"
    elif 0 <= idade <= 5:
        return "0-5 anos"
    elif 6 <= idade <= 12:
        return "6-12 anos"
    elif 13 <= idade <= 17:
        return "13-17 anos"
    return "Fora do Escopo"

df['faixa_etaria'] = df['idade'].apply(calcular_faixa_etaria)

# Extrair ano
df['ANO'] = df['AP_CMP'].astype(str).str[:4]
```
